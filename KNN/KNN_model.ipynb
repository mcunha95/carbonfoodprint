```python
import numpy as np
from sklearn.preprocessing import Imputer
from sklearn.cross_validation import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
import pandas as pd
import random
```

```python
data_csv = pd.read_csv("./formated_normalized_food_per_country_ageGroup.csv")
data_csv
```

```python
X = np.asarray(data_csv.iloc[:,2:])
y_country = np.asarray(pd.get_dummies(data_csv.iloc[:,0]))
y_age = np.asarray(pd.get_dummies(data_csv.iloc[:,1]))
X.shape
```

```python
neigh_country = KNeighborsClassifier(n_neighbors = 5, weights='uniform', algorithm='auto')
neigh_country.fit(X, y_country)
neigh_age = KNeighborsClassifier(n_neighbors = 1, weights='uniform', algorithm='auto')
neigh_age.fit(X, y_age)
```

```python
i = 165#random.randint(0,len(data_csv))
y_pred_country = neigh_country.predict(X[i].reshape(1, -1))
y_pred_country = [country for country,index in zip(list(pd.get_dummies(data_csv.iloc[:,0]).columns.values),y_pred_country.tolist()) if index][0] # Converting from one hot encoding to country name
y_actual_country = [country for country,index in zip(list(pd.get_dummies(data_csv.iloc[:,0]).columns.values),y_country[i].tolist()) if index][0] # Converting from one hot encoding to country name
print(X[i].reshape(1, -1))
print('Predicted country: ' + y_pred_country)
print('Actual country: ' + y_actual_country)
```
