```python
import pandas as pd
```

```python
food_data = pd.read_csv('./df/CleanFoodData.csv', sep=',')
food_data.rename(columns={'Mean': 'Mean_Food.g'}, inplace=True)
food_data['Mean_CO2.g'] = [0] * len(food_data) # Initialize column for CO2
food_data.head(3)
```

```python
co2_data = pd.read_csv('./df/CO2Footprint.csv', sep=',', index_col = 0)
co2_data['Grams.CO2e.per.Grams'] = co2_data['Grams.CO2e.per.Serving']/co2_data['Serving.Size'] # Created column for CO2/grams of food
co2_data.head(3)
```

```python
for iFood, foodRow in food_data.iterrows():
    for co2Id, co2Row in co2_data.iterrows():
        if co2Id == foodRow['ID']:
            food_data.loc[iFood,'Mean_CO2.g'] = foodRow['Mean_Food.g']*co2Row['Grams.CO2e.per.Grams']
            break
food_data
```
